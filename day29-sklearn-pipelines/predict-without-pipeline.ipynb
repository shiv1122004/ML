```python
# Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Import necessary libraries (you'll need these in the new notebook)
import pickle
import os

# Define the path to your models in Google Drive
models_path = '/content/drive/MyDrive/colab_models'

# Check if the directory exists and list its contents
if os.path.exists(models_path):
    print(f"Models found in: {models_path}")
    print(f"Files: {os.listdir(models_path)}")

    # Load your models
    ohe_sex = pickle.load(open(os.path.join(models_path, 'ohe_sex.pkl'), 'rb'))
    ohe_embarked = pickle.load(open(os.path.join(models_path, 'ohe_embarked.pkl'), 'rb'))
    clf = pickle.load(open(os.path.join(models_path, 'clf.pkl'), 'rb'))

    print("Models loaded successfully!")
else:
    print(f"Error: Models directory not found at {models_path}")
```

```python
# Assume user input
# Pclass/gender/age/SibSp/Parch/Fare/Embarked
test_input = np.array([2, 'male', 31.0, 0, 0, 10.5, 'S'],dtype=object).reshape(1,7)
```

```python
test_input
```

```python
test_input_sex = ohe_sex.transform(test_input[:,1].reshape(1,1))
```

```python
test_input_sex
```

```python
test_input_embarked = ohe_embarked.transform(test_input[:,-1].reshape(1,1))
```

```python
test_input_embarked
```

```python
test_input_age = test_input[:,2].reshape(1,1)
```

```python
test_input_transformed = np.concatenate((test_input[:,[0,3,4,5]],test_input_age,test_input_sex,test_input_embarked),axis=1)
```

```python
test_input_transformed.shape
```

```python
clf.predict(test_input_transformed)
```
